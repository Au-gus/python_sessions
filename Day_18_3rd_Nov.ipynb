Local Variables:
        The variables provided inside the function.
Global Variables:
        The variables provided outside the fucntion.

```python
# Local Variables
def add():
    n1=10
    n2=20
    a=n1+n2
    print("the addition is {}".format(a))
add()
```

```python
# Global Variables
n1=10
n2=20
def add():
    a=n1+n2
    print("the addition is {}".format(a))
add()
```

```python
# using local value outside the function without return statement
n1=10
def up():
    global n1
    n1=20
    n1=n1+100
    print("n1 value inside function is:",n1)
up()
print(n1)

# we need updated value without returning the value
```

```python
def up():
    global n11,n22
    n11=20
    n22=30
    print(n11)
up()
print(n11)
print(n22)
```

```python
a = eval(input("Enter a number:"))
b = eval(input("Enter a number:"))
def add(a,b):
    print(a+b)

def sub(a,b):
    print(a-b)

def mul(a,b):
    print(a*b)

def div(a,b):
    print(a/b)

def main():
    operation = eval(input("Enter your desired number for operation:"))
    if (operation == 1):
        add(a,b)
    elif (operation == 2):
        sub(a,b)
    elif (operation == 3):
        mul(a,b)
    elif (operation == 4):
        div(a,b)
    else:
        print("Error")

main()
    
    
```
